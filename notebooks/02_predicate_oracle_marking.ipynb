# QDSV Bridge Colab 02 - Predicate Oracle Marking

This notebook uses the `predicate_marking` family to derive oracle-ready artifacts from a bounded predicate-oriented problem specification.

```python
import os
from IPython.display import Markdown, display
from qdsv_bridge import QDSVBridgeClient

API_URL = os.getenv("QDSV_BRIDGE_API_URL", "https://qdsv-api-568788785178.us-central1.run.app/api")
API_KEY = os.getenv("QDSV_BRIDGE_API_KEY") or None
client = QDSVBridgeClient(api_url=API_URL, api_key=API_KEY, timeout=60)
```

```python
spec = {
    "family": "predicate_marking",
    "bridge_mode": "build",
    "state_space": {
        "kind": "finite_candidates",
        "candidate_count": 32,
        "candidate_id": "item",
    },
    "signals": ["capacity_ok", "risk_band", "priority_score"],
    "goal": {
        "kind": "marking",
        "predicate": "capacity_ok AND risk_band IN allowed_range AND priority_score >= threshold",
    },
    "target": {
        "format": "qiskit_blueprint",
        "backend_family": "qiskit",
    },
    "limits": {
        "max_qubits": 7,
        "max_depth": 300,
    },
}

artifact = client.build(spec)
print("status:", artifact["status"])
print("family:", artifact["family"])
print("artifact format:", artifact["artifact"]["format"])
print("semantic preservation:", artifact["semantic_preservation"])

print("\nQiskit blueprint preview:\n")
print(artifact["artifact"].get("content", "")[:1600])
```

```python
report = client.report(spec, mode="build", format="markdown")
display(Markdown(report["content"]))

with open("bridge_report_predicate_oracle_marking.md", "w", encoding="utf-8") as f:
    f.write(report["content"])

print("Saved: bridge_report_predicate_oracle_marking.md")
```
